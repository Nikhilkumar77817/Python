#**Loops in Python - For, While and Nested Loops**





* Loops in Python are used to repeat actions efficiently. The main types are For loops (counting through items) and While loops (based on conditions).

**For Loop**
* For loops is used to iterate over a sequence such as a list, tuple, string or range.
* It allow to execute a block of code repeatedly, once for each item in the sequence.

**While Loop**
* In Python, a while loop is used to execute a block of statements repeatedly until a given condition is satisfied.
* When the condition becomes false, the line immediately after the loop in the program is executed.

**Nested Loops**
* Python programming language allows to use one loop inside another loop which is called nested loop.

```python
#Palindrome number
n=int(input("Enter number: "))
temp=n
rev=0
while n>0:
    rev=rev*10+(n%10)
    n//=10
if rev==temp:
    print(" yes ,it is Palindrome")
else:
    print("Not Palindrome")
```

**Continue And Break Statement**
* The **continue** statement in Python returns the control to the beginning of the loop.
* The **break** statement in Python brings control out of the loop.

```python
for i in range(5):
    if i == 2:
        continue
    print(i)
```

```python
i = 0
while i < 5:
    print(i)
    if i == 2:
        break
    i += 1
```

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
  if x == "banana":
    break
```

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  if x == "banana":
    continue
  print(x)
```

```python
rows = int(input("Enter number of rows: "))

k = 0

for i in range(1, rows+1):
    for space in range(1, (rows-i)+1):
        print(end="  ")

    while k!=(2*i-1):
        print("* ", end="")
        k += 1

    k = 0
    print()
```

```python
#  Pascal's Triangle

rows = int(input("Enter number of rows: "))
coef = 1

for i in range(1, rows+1):
    for space in range(1, rows-i+1):
        print(" ",end="")
    for j in range(0, i):
        if j==0 or i==0:
            coef = 1
        else:
            coef = coef * (i - j)//j
        print(coef, end = " ")
    print()
```

```python
# Floyd's Triangle

rows = int(input("Enter number of rows: "))
number = 1

for i in range(1, rows+1):
    for j in range(1, i+1):
        print(number, end=" ")
        number += 1
    print()
```
